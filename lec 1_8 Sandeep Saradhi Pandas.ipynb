## Different ways of creating Series in pandas

# Series
 1. Empty Series

```python
import pandas as pd
s=pd.Series() # create an empty series --> no parameters here.
```

```python
# to work with arrays we need to use numpy here.
import numpy as np
ar1=np.array([1,2,3,4,5,7,7,3,4,5,2])
s1=pd.Series(ar1)
print(s1)
```

```python
# we can also use index parameter here. default value starts with 0 only. we have to pass indexes as lists here.
import numpy as np
ar1=np.array([1,2,3,4,5])
s1=pd.Series(ar1,index=['a','b','c','d','e'])
print(s1)
```

```python
dict={'a':10,'b':20,"c":30,'d':40}
pd.Series(dict)
```

```python
# instead of creating a dictionary,we can also create a list of tuples here.
t1=[("a",10),('b',200),("c",100),('d',400)] # here the complete tuple will be treated as single elements.
print(pd.Series(t1))
```

# lec 5
## Attributes of Series in Pandas

1. Index: -> Series.index --> Returns all index values.
2. array: -> Series.array --> returns an array of values.
3. values: ->Series.values --> returns the value of series
4. name: -> Series.name  --> returns name of series if it is present else empty result.
5. shape: ->Series.shape --> returns the shape of the series
6. ndim:  -> Series.ndim --> returns the dimension of the series
7. size:  -> Series.size --> returns the size of the series
8. nbytes: -> Series.nbytes --> returns the memory occupied by only values
9. memory_usage: Series.memory_usage --> returns the total memory occupied by both indexes and vales in series.
10. empty: -> Series.empty --> returns True if the series is empty.
    

```python
import pandas as pd
s1=pd.Series([1,2,3,4,5,3,2,1,5,6,7,5,4,3,22,5,6,7,8,94,3],name="Series1")
print(s1)
print("Indexes are:",s1.index)
print()
print("Name of the series is:",s1.name)
print("Shape of series is:",s1.shape)
print()
print("Array values are:",s1.array)
print()
print("Values are:",s1.values)
print("Dimension of the series is:",s1.ndim)
print("Size of the series is:",s1.size)
print("Memory occupied by values is:",s1.nbytes)
print("Memory for both indexes and values is:",s1.memory_usage())
print("Is series Empty or not?", s1.empty)
```

```python
print(s1.memory_usage(index=False)) # without indexes
```

```python
s1.memory_usage()
```

```python
s1.dtype
```

```python
pd.Series().empty
```

# lec 6
## Mathematical functions on Series in Pandas

```python
#add,sub,div,mul,less than ,greater than,mod,power,divide
# to perform these mathematical operations we need to create two series.
```

```python
s1=pd.Series([1,2,3,5,6,7,9])
s2=pd.Series([2,3,4,5,6,7,8])
print(s1)
print(s2)
```

```python
# addition # we will perform both two ways here.
print(s1+s2)

print()

print(s1.add(s2))
```

```python
pd.add(s1,s2) # we cannot perform like this.
```

```python
# subtraction
print(s1-s2)
print()
print(s1.subtract(s2))
```

```python
# multiplication
print(s1*s2)
print()
print(s1.multiply(s2))
```

```python
# division
print(s1/s2)
print()
print(s1.divide(s2))
```

```python
# power
print(s1**s2)
print()
print(s1.pow(s2))
```

```python
# modulo
print(s1%s2)
print()
print(s1.mod(s2))
```

```python
# less than 
print(s1<s2)
print()
print(s1.lt(s2))
```

```python
# less than  or equal to
print(s1<=s2)
print()
print(s1.le(s2))
```

```python
# greater than  or equal to
print(s1>=s2)
print()
print(s1.ge(s2))
```

```python
# greater than  
print(s1>s2)
print()
print(s1.gt(s2))
```

```python
# equal to
print(s1==s2)
print()
print(s1.eq(s2))
```

### sort_values(column_name) is used to sort the data either in ascending or the descending order.

```python
# to get the Maths marks in descending order,we need to pass one more argument as ascending=False.
df1.sort_values('Maths',ascending=False) # based on Maths marks we are performing the sorting operation here.
```

```python
df1.sort_values('Maths')  # by default ascending value will be always True only.
```

```python
df1.sort_values('Names',ascending=False)
```

```python
df1.sort_values('Names') # sorting based on the names column
```

#### to sort two diiferent columns we need to pass them as a list

```python
df1.sort_values(['Names','Maths','Social']) # first mainly sorting will be performed on the Names .
```

```python
df1.sort_values(['Names','Maths','Social'],ascending=False)
```

```python
df1.sort_values(['Maths','Names','Social']) # maths marks have high priority here.
```

### we can also give which one need to be ascending and which column have to be descending here.

```python
# here names are ascending, maths should be descending so 1 for it and Social also 1 since we want descending here.
df1.sort_values(['Names','Maths','Social'],ascending=[0,1,1]) # 0 represents Descending and 1 represents Ascending here.
```
