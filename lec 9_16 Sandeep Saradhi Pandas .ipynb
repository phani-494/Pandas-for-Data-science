## Understanding loc[] and iloc[] in Pandas

```python
import pandas as pd
data=pd.read_excel("C:\\Users\\Phanindra Reddy\\OneDrive\\Desktop\\GATE NEW PLAN\\GATE TIME TABLE.xlsx")
df2=pd.DataFrame(data)
print(df2)
```

```python
print(df2)
```

```python
# loc is completely based on labels.
# df2.loc[rows,columns] # here rows will be a number,whereas the columns will be their names.
df2.loc[3,'Maths'] # we need to give the name of the column here.
```

```python
df2.loc[3,['English','Maths']]
```

```python
df2.loc[3] # entire 3rd row we get here.
```

```python
df2.loc[0:8] # this will include the index 8 also which means 9 elements will present here.
```

```python
df2.loc[0:8,'Names']  # if u want only 0:8 rows with their names only,then we can give like this.
```

```python
# to get the multiple columns here,we need to give the list here in second parameter.
df2.loc[0:8,['Names','Social']]
```

```python
# we can also use the slicing process here which means from this column name to this column name.
df2.loc[5:8,'Names':'Social'] # from names column to social column will be considered.
```

```python
df2.iloc[2,3] # which is second row and third column element.
```

```python
df2.iloc[3:10,0:3] # when we use iloc, we won't get the last value. 
```

```python
df2.iloc[3:10,0]
```

```python
# suppose if we want only any  2 rows ,then we need to give in list 
df2.iloc[[0,5]]
```

```python
df2.iloc[:,[0,2]] # only 0 and 2 column is displayed here.all rows will be displayed here.
```

## Handling missing data (dropna & fillna) in Pandas

1. remove the missing data, which means completely remove all the rows which doesn't have any data or missed the data. -->dropna
2. fill the data with default values or the values which we want  --> fillna

```python
import pandas as pd
data=pd.read_excel("C:\\Users\\Phanindra Reddy\\OneDrive\Desktop\\GATE NEW PLAN\\New Marks.xlsx")
d1=pd.DataFrame(data)
print(d1)
# wherever data is not present,at all those places NaN is present .
```

```python
d1.dropna() # rows with NaN will be deleted here. # creates the new copy here. original doesn't affect here.
```

```python
print(d1)
```

```python
# if we want the changes to take place on the original dataframe, then we need to give a parameter inplace as True.
d1.dropna(inplace=True) # it won't give any output here.
print(d1)
```

```python
# suppose we want to drop  NaN values only in a column,then we can also perform it using dropna.
# just rerun the first command to get all the original data back.
d1['Maths']
```

```python
d1['Maths'].dropna()
```

```python
# suppose we want to drop the data in multiple columns then we can use the loc here
```

```python
d1.loc[:,['Maths','English']].dropna() # we can give the inplace parameter to get changes in the original array.
```

# fillna()

```python
# we want to fill the missing data values with any default value or any some values.so we can use fillna here.
d1.fillna('500') # see we get 500 here inn all the missing places.
# here also the original dataframe will not be affected. so we need to give the parameter inplace here.
```

```python
# for the English here we want to fill the default value wherever data is missed in English column.
# the original data frame is still not affected here. so we can perform this operation here.
d1['English'].fillna(175)
```

```python
d1.loc[:,['Maths','English']].fillna(1000)
```

```python
# we want to replace the missing values with average here. use mean() to calculate average
d1['English'].fillna(d1['English'].mean())
```

```python
import math
d1['English'].fillna(math.floor(d1['English'].mean()))
```

## Data Filtering (using conditions) in Pandas 

```python
data=pd.read_excel("C:\\Users\\Phanindra Reddy\\OneDrive\\Desktop\\GATE NEW PLAN\\New Marks_without index.xlsx")
df4=pd.DataFrame(data)
print(df4)
```

#### in loc we will give the condition and retrieve the data based on the condition.

```python
df4.loc[df4['English']<200] # students who get less than 200
```

```python
df4.loc[0,'English']
```

```python
 # we can also give the multiple conditions here.
df4.loc[(df4['English']<200) &  (df4['Maths']<165)]
```

```python
df4.loc[(df4['English']<200) |  (df4['Maths']<165)] # here we can give the multiple conditions.
```

```python
# we can also perform the slicing operation here based on the strings.
df4.loc[df4["Names"].str.contains("an")]
# returns the strings which all have the "an" substring
```

```python
df4.loc[df4["Names"].str.contains("ant")] # empty since no string contains it.
```

```python
# when we give conditions,we should not write and /or . we should use the logial symbols.
df4.loc[(df4['English']<200) and  (df4['Maths']<165)] 
```

```python
# we want all the elements which doesn't contain the string part then use ~ operator.
df4.loc[~df4["Names"].str.contains("ant")]
```

```python
# we want a data which starts with some string. we can use startswith()
df4.loc[df4["Names"].str.startswith("i")]
```

```python
df4.loc[~df4["Names"].str.startswith("i")] # which doesn't start with i
```

```python
df4.loc[df4["Names"].str.endswith("i")]
```

```python
df4.loc[~df4["Names"].str.endswith("i")] # which doesn't ends wih i
```

## Conditional changes in Pandas

```python
# based on a condition,we can change the row or a column values 
```

```python
import pandas as pd
data=pd.read_excel("C:\\Users\\Phanindra Reddy\\OneDrive\\Desktop\\GATE NEW PLAN\\GATE TIME TABLE.xlsx")
df2=pd.DataFrame(data)
print(df2)
```

#### here we will create a percentage column and based on the percentage ,we will decide who is pass or fail.

```python
df2['Total']=df2['English']+df2['Maths']+df2['Social']
```

```python
df2
```

```python
df2['Percentage']=(df2['Total']/600)*100
```

```python
df2
```

```python
# first we will initialize the grade with none here.
df2['Grade']='None'
print(df2)
```

```python
df2.loc[df2['Percentage']<90,['Grade']]="FAIL"
print(df2)
# here we are changing the column values based on a condition.
```

```python
df2.loc[df2['Percentage']>=90,['Grade']]="PASS"
print(df2)
```
